```python
import math
import numpy as np
import matplotlib.pyplot as plt


def logistic(X):
    return (1 + np.tanh(X))/2

def pointslope(m,bias,X):
    return (2**m) * X + bias

def intercept(X):
    return logistic(X)-pointslope(m,bias,X)

m1=-8
X1 = np.arange(-4, 4, 0.2)
X2 = np.arange(-2,4,0.2)
X3 = np.arange(0, 4, 0.2)
X4 = np.arange(2,4,0.2)

f1 = plt.figure(figsize=(20,10))
plt.subplot(221)
plt.plot(X1 , logistic(X1),linewidth=3,linestyle='-.')
for i in range(2):
    bias = logistic(min(X1))-2**m1*min(X1)
    plt.plot(X1 , pointslope(m1,bias,X1),label=f"m={m1}",linewidth=3)
    m1 += 1
    plt.legend(fontsize=18)
plt.ylabel("Probability",fontsize=20)
plt.xlabel("Variable X",fontsize=20)
plt.grid(True)

m2 = -8
plt.subplot(222)
plt.plot(X1 , logistic(X1),linewidth=3,linestyle='-.')
for i in range(7):
    bias = logistic(min(X2))-2**m2*min(X2)
    plt.plot(X2 , pointslope(m2,bias,X2),label=f"m={m2}",linewidth=3)
    m2 += 1
    plt.legend(fontsize=18)
plt.ylabel("Probability",fontsize=20)
plt.xlabel("Variable X",fontsize=20)
plt.grid(True)

m3 = -8
plt.subplot(223)
plt.plot(X1 , logistic(X1),linewidth=3,linestyle='-.')
for i in range(8):
    bias = logistic(min(X3))-2**m3*min(X3)
    plt.plot(X3 , pointslope(m3,bias,X3),label=f"m={m3}",linewidth=3)
    m3 += 1
    plt.legend(fontsize=18)
plt.ylabel("Probability",fontsize=20)
plt.xlabel("Variable X",fontsize=20)
plt.grid(True)

m4 = -8
plt.subplot(224)
plt.plot(X1 , logistic(X1),linewidth=3,linestyle='-.')
for i in range(2):
    bias = logistic(min(X4))-2**m4*min(X4)
    plt.plot(X4 , pointslope(m4,bias,X4),label=f"m={m4}",linewidth=3)
    m4 += 1
    plt.legend(fontsize=18)
plt.ylabel("Probability",fontsize=20)
plt.xlabel("Variable X",fontsize=20)
plt.grid(True)
```
